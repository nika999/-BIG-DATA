В результаті виконання лабораторної роботи необхідно відповісти на запитання:
1. Виведіть перші 6 назв фільмів дата фрейму.

3. Скільки фільмів мають тривалість менше 100 хв.


```python
import pandas as pd
import urllib3
import bs4
from bs4 import BeautifulSoup
http = urllib3.PoolManager()
r = http.request('GET', 'https://www.imdb.com/search/title/?count=100&release_date=2017,2017&title_ty')
r.status
# 200 HTTP Status Codes means OK
```

```python
d = r.data
soup = BeautifulSoup(d, 'html.parser')
sp =soup.prettify()
```

```python
title =[]
for link in soup.find_all("h3", class_="lister-item-header"):
    
    name = link.get_text()
    title.append(name)
```

```python
runtime =[]
for link in soup.find_all('span', class_="runtime"):
    name2 = link.get_text()
    runtime.append(name2)
```

```python
title = list(map(str.strip,title))
runtime = list(map(str.strip,runtime))
```

```python
df = pd.DataFrame({'Title':title,'Runtime':runtime})
df
```

```python
spec_chars = ["\n","(", ")", ".", "–"]
for char in spec_chars:
    df['Title'] = df['Title'].str.replace(char, ' ')
    
df['Title'] = df['Title'].str.replace('\d+', '')
df
```

### 2. Виведіть всі назви фільмів с тривалістю більше 120 хв.

```python
df['Runtime'] = df['Runtime'].map(lambda x: x.lstrip('+-').rstrip('min'))
```
